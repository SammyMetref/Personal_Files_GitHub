```python
# OSMOSIS: SSH from NATL60, daily outputs, degradated grid (3x3)
file_deg='/Users/sammymetref/Documents/Boost-Swot/Notebooks/GitHub/Personal_Files/2018/Data/DA_OSMOSIS/QGSWOSMO_IniNATL60_Ensforecast/QGSWOSMO_IniNATL60_Ensforecast_y2012m9d1h0_SSHdegrad.nc'
fid_deg = nc.Dataset(file_deg)
lon2d=np.array(fid_deg.variables["lon"][:,:]) 
lat2d=np.array(fid_deg.variables["lat"][:,:])  
SSH_NATL60_degrad=np.array(fid_deg.variables["SSH"][:,:,:]) 
  
```

```python
i_ens=0 

plt.figure(figsize=(20, 10))

plt.subplot(131)
plt.pcolormesh(lon2d,lat2d,np.mean(SSH_NATL60_degrad[:,:,:],0))
plt.colorbar(extend='both', fraction=0.042, pad=0.04)
plt.title('ensemble mean SSH');

plt.subplot(132)
plt.pcolormesh(lon2d,lat2d,SSH_NATL60_degrad[i_ens,:,:])
plt.colorbar(extend='both', fraction=0.042, pad=0.04)
plt.title('SSH i_ens=1');

plt.subplot(133)
plt.pcolormesh(lon2d,lat2d,SSH_NATL60_degrad[i_ens+1,:,:])
plt.colorbar(extend='both', fraction=0.042, pad=0.04)
plt.title('SSH i_ens=2');
 
```
